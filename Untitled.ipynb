```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(binary=True)
corpus = ["Tessaract is good optical character recognition engine","optical character recognition is significant"]
vect.fit(corpus)

vocab = vect.vocabulary_
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
    
print(vect.transform(["This is a good optical illusion. It is a good one."]).toarray())

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vect.transform(["Google Cloud Vision is a character recognition engine"]))
print(similarity)

print(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray())
```

```python
def gender_features(word):
    return {'last_letter':word[-1]}
```

```python
gender_features('Kirti')
```

```python
import nltk
from nltk.corpus import names
```

```python
names.words()
```

```python
print(len(names.words()))
```

```python
labeled_names = [(name,'male') for name in names.words('male.txt')] + [(name,'female') for name in names.words('female.txt')]
```

```python
import random
random.shuffle(labeled_names)
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labeled_names]
```

```python
train_set,test_set = featuresets[:5000], featuresets[2000:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Gayathri'))
```

```python
classifier.classify(gender_features('Kannan'))
```

```python
classifier.classify(gender_features('Sumanraj'))
```

```python
classifier.classify(gender_features('Samantha'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```
